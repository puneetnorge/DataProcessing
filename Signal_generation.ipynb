```python
import numpy as np
import matplotlib.pyplot as plt

## Author: Puneet Sharma as part of Fundamentals of Data Processing

## Generate several copies of a basic signals with different levels/types of noise and missing values i.e., NaNs

# Define signal parameters
t = np.linspace(0, 1*np.pi, num=1000, endpoint=True)  # Time vector
## Two sine waves with different frequencies 
f1 = 2  # Frequency of the first sine wave
f2 = 3  # Frequency of the second sine wave
amp1 = 1  # Amplitude of the first sine wave
amp2 = 0.75  # Amplitude of the second sine wave
sigma = 0.3  # sigmal of Gaussian noise
mu = 0 # mean in Gaussian noise
missing_data_per = 0.2  # Percentage of missing data

# Generate signals s1 and s2 and add them to create a new mixed signal
s1 = amp1 * np.sin(2 * np.pi * f1 * t)
s2 = amp2 * np.sin(2 * np.pi * f2 * t)
mixed_signal = s1 + s2;


## Noise with gaussian or normal distribution
noise_gauss = np.random.normal(mu, sigma, len(t)) # mean, standard deviation, and size
plt.figure(figsize=(10, 5))
plt.hist(noise_gauss, 20, density=True, histtype='bar', rwidth=0.8)
plt.show()
mixed_signal_gauss = mixed_signal + noise_gauss

## Uniform noise is uniformly distributed across a specified range
low = np.min(mixed_signal)/4
high = np.max(mixed_signal)/4
noise_uniform = np.random.uniform(low, high, len(t))
mixed_signal_uniform = mixed_signal + noise_uniform



## Impulse (Salt-and-Pepper) Noise: consists of occasional sudden increase or drop in intensity
per_of_noise = 10; # percentage of data to be injected with noise
num_indices = int(per_of_noise*0.01 * len(t))
# Generate random indices
random_indices = np.random.choice(len(t), size=num_indices, replace=False)
mixed_signal_impulse = mixed_signal
# Calculate the number of indices to select
mixed_signal_impulse[random_indices] =  np.random.uniform(np.min(mixed_signal),np.max(mixed_signal),len(random_indices))

# Missing data is defined as NaN
missing_data_indices = np.random.choice(len(t), int(len(t) * missing_data_per), replace=False)
mixed_signal_with_missing_data = mixed_signal.copy()
mixed_signal_with_missing_data[missing_data_indices] = np.nan # replace selected indices with NaNs



# Plot the different noisy signals
plt.figure(figsize=(10, 5))
plt.plot(t, mixed_signal_gauss, label='Mixed Signal with Guassian Noise')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(t, mixed_signal_uniform, label='Mixed Signal with Uniform Noise')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

# Plot the different noisy signals
plt.figure(figsize=(10, 5))
plt.plot(t, mixed_signal_impulse, label='Mixed Signal with Impulse Noise')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 5))
plt.plot(t, mixed_signal_with_missing_data, label='Mixed Signal with Missing values')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

```
