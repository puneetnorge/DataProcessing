```python
# Create an list of dictionaries
people = [
    {
        "name": "Jane Seigmann",
        "age": 35,
        "city": "Alta"
    },
    {
        "name": "John Nordmann",
        "age": 40,
        "city": "Tromsø"
    },
    {
        "name": "Joe Johnson",
        "age": 28,
        "city": "Bergen"
    }
]
# Access values from the first dictionary
print("Name of first person:", people[0]["name"])  
print("City of second person:", people[1]["city"])  
# Add a new key-value pair to the first dictionary
people[0]["job"] = "Engineer"
print("Updated First Dictionary:", people[0])

# Update an existing value in the second dictionary
people[1]["age"] = 41
print("Updated Age of Second Person:", people[1]["age"])  # Output: 41
# Remove a key-value pair from the third dictionary
del people[2]["city"]
print("Third Dictionary After Deletion:", people[2])

# Loop through the array of dictionaries
print("People Information:")
for person in people:
    for key, value in person.items():
        print(f"{key}: {value}")
    print("---")  # Separator for each person
keys = people[0].keys()
values = people[0].values()
print("Keys in First Dictionary:", keys)   
print("Values in First Dictionary:", values)
```
