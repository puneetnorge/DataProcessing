```python
import scipy.linalg as la
import numpy as np

# Define a system Ax = b
A = np.array([[1, 3], [4, 10]])
b = np.array([2, 4])

print(f"A = {A}")
print(f"b = {b}")
# Solve for x
x = la.solve(A, b)
print(f"Solution x = {x}")
```

```python
from scipy.optimize import minimize

# function to minimize
def func(x):
    return (x - 5) ** 2 + 7

# Initial guess
x0 = [0]

# Minimize the function
result = minimize(func, x0)
print("Minimum Value of f(x):", result.fun)
print("Optimal x:", result.x)
```

```python
from scipy.interpolate import interp1d


# Given data points
x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 3, 7, 13, 21])

# Create linear interpolation function
f_inter = interp1d(x, y, kind='linear')

# Interpolate value at 
x = 2.5
print(f"Interpolated value of f_inter(x) at x = 2.5 = {f_inter(x)}")
```
