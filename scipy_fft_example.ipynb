```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Parameters
fs = 1000  # Sampling frequency in Hz
T = 1.0    # Duration in seconds
N = int(fs * T) # Number of samples should be an integer
t = np.linspace(0, T, int(N), endpoint=False)  # Time vector

# Generate a signal wave
freq_sin = 50  # Frequency of the sine wave in Hz
freq_cos = 25  # Frequency of the sine wave in Hz
signal = np.sin(2 * np.pi * freq_sin * t) + 2*np.sin(2 * np.pi * freq_cos * t) 

# plot only a first tens of samples of the signal
no_of_samples_to_plot = 50
plt.figure(figsize=(8, 4))
plt.plot(t[0:no_of_samples_to_plot],signal[0:no_of_samples_to_plot])
plt.grid()
plt.xlabel("Time (in sec)")
plt.ylabel("Amplitude")

# Compute FFT
fft_values = fft(signal)
fft_magnitude = np.abs(fft_values) / N  # Normalize
frequencies = fftfreq(N, 1/fs)  # Frequency bins

# Plot FFT
plt.figure(figsize=(8, 4))
plt.plot(frequencies, fft_magnitude)  # TWO-sided spectrum
plt.title("FFT of Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid()
plt.show()
```
