```python
import random
"""Sensor Class with methods"""
class Sensor:
    """This is a Initilization method (sort of Constructor)"""
    def __init__(self, temperature, relative_humidity, co2_concentration, model_id) -> None:
        # -> None indicates that this method returns nothing
        self.temperature = temperature
        self.relative_humidity = relative_humidity
        self.co2_concentration = co2_concentration
        self.model_id = model_id
    
    """This method is used to display the variables of Sensor Object"""
    def display_parameters(self) -> None:
        print(" Model ID "+self.model_id+"\n Temp in (C) "+str(self.temperature)
             +"\n Relative Humidity(in perc.)"+str(self.relative_humidity)+ 
             "\n CO2 Conc. "+str(self.co2_concentration))
        
        
    """Methods used to change the variables"""
    def change_temp(self, temperature) -> None:
        self.temperature = temperature
    def change_relHum(self, relative_humidity) -> None:
        self.relative_humidity = relative_humidity
    def change_CO2_conc(self, co2_concentration) ->None:
        self.co2_concentration = co2_concentration
    
  

    
    

```

```python
# initalize a sensor object with some values
obj_sen = Sensor(22.04,33.044,200,"GTX2023") 
random.seed(5)
print("100 random values associated with a Sensor Object")
for i in range(100):
    c1 = random.randint(10,2000) 
    h1 = random.random()*100;
    t1 = random.uniform(-10,50);
    obj_sen.change_temp(t1)
    obj_sen.change_relHum(h1)
    obj_sen.change_CO2_conc(c1)
    obj_sen.display_parameters()
    

```
